```python
import numpy as np
import matplotlib.pyplot as plt
# data generation
n=10
X = np.linspace(-10,10,n)
Y = 25 + 0.5*X**2 + 0.3*X -0.042*(X**3)
factor = 15
Y = (0.1*(X+4)**2 - 0.07*(X**3) - 20)/factor + (np.random.randint(-10,10,n))/10
xWanted = np.linspace(-10,10,100)
yWanted = (0.1*(xWanted+4)**2 - 0.07*(xWanted**3) - 20)/factor
# plot
plt.scatter(X, Y)
plt.scatter(xWanted,yWanted, color="red", alpha=0.2)
plt.show()
```

```python
features = np.array([lambda _: 1,
                     lambda x: x,
                     lambda x: x**2,
                     lambda x: x**3,
                     lambda x: x**4,
                     lambda x: x**5,
                     lambda x: x**6])
```

```python
epochs = 100000000000
alpha = 0.0000000000009
W = np.zeros(len(features))
for epoch in range(epochs):
    i=0
    gradient = []
    for f in features:
        pred = np.array([sum([w*f(x) for f,w in zip(features,W)]) for x in X])
        error = Y - pred
        phi_x = np.array([f(x) for x in X])
        d = np.sum(phi_x * error)
        gradient.append(d)
        d *= alpha
        W[i] += d
        i+=1
    norm = np.linalg.norm(gradient)
    if epoch%2500 == 0: print("\rdone", epoch, "epochs,  norm:", norm, end = "")
print(W)
```

```python
epochs = 100000000000
oldEpochs = epoch
alpha = 0.0000000000009
for epoch in range(epochs):
    i=0
    gradient = []
    for f in features:
        pred = np.array([sum([w*f(x) for f,w in zip(features,W)]) for x in X])
        error = Y - pred
        phi_x = np.array([f(x) for x in X])
        d = np.sum(phi_x * error)
        gradient.append(d)
        d *= alpha
        W[i] += d
        i+=1
    norm = np.linalg.norm(gradient)
    if epoch%2500 == 0: print("\rdone", epoch+oldEpochs, "epochs,  norm:", norm, end = "")
print(W)
```

```python
Z = [sum([w*f(x) for w,f in zip(W,features)]) for x in X]
plt.scatter(X, Z, color="red")
plt.scatter(X, Y)
plt.show()
```
