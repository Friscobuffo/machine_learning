```python
import numpy as np
import matplotlib.pyplot as plt
# data generation
n=10
X = np.linspace(-10,10,n)
Y = 25 + 0.5*X**2 + 0.3*X -0.042*(X**3)
factor = 15
Y = (0.1*(X+4)**2 - 0.07*(X**3) - 20)/factor + (np.random.randint(-10,10,n))/10
xWanted = np.linspace(-10,10,100)
yWanted = (0.1*(xWanted+4)**2 - 0.07*(xWanted**3) - 20)/factor
# want all column vectors
X = np.reshape(X, (n,1))
Y = np.reshape(Y, (n,1))
xWanted = np.reshape(xWanted, (100,1))
yWanted = np.reshape(yWanted, (100,1))
# plot
plt.scatter(X, Y)
plt.scatter(xWanted,yWanted, color="red", alpha=0.2)
plt.show()
```

```python
features = np.array([lambda _: 1,
                     lambda x: x,
                     lambda x: x**2,
                     lambda x: x**3,
                     lambda x: x**5,
                     lambda x: x**6,
                     lambda x: x**7])
```

```python
alpha  = 0.0000000000012
# W = np.ones((len(features),1))
epochs = 500000
for _ in range(epochs):
    phi = np.matrix([[f(x) for f in features] for [x] in X])
    phi_w = np.matmul(phi, W)
    error = Y - phi_w
    gradient = -np.matmul(phi.T, error)
    W -= alpha * gradient
print(W)
print(np.linalg.norm(gradient))
```
