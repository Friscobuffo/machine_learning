```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

plt.style.use('_mpl-gallery')
n = 40
# Make data
X = np.linspace(0, 5, n)
Y = np.linspace(0, 5, n)
X, Y = np.meshgrid(X, Y)
Z = np.sqrt(X**2 + Y**2)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
print(Z.shape)
print(X.shape)
```
