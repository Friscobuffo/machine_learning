```python
import numpy as np
import matplotlib.pyplot as plt
from math import ceil
from random import random
n=40
X = np.random.randint(30, 250, (n,1)) # square meters
Y = np.reshape([ceil(x) for [x] in ((X-30) / 220) * 3], (n,1))
def rumor(x):
    if x == 1:
        return 1
    r = random()
    if x == 2:
        if r<0.3: return 1
        return 2
    if r<0.5: return 2
    return 3
Y = np.reshape([rumor(y) for [y] in Y], (n,1)) # number of bathrooms
Z = 2500*X + 30000*(Y**2) # price
Z = np.reshape([z+(0.5-random())*80000 for [z] in Z], (n,1))
```

```python
b1,b2,b3 = 0,0,0
for [y] in Y:
    if y==1: b1+=1
    if y==2: b2+=1
    if y==3: b3+=1
print(b1,b2,b3)
fig, ax = plt.subplots(1,2, figsize=(10, 3))
ax[0].scatter(X,Z)
ax[0].set_xlabel("Square Meters")
ax[0].set_ylabel("House Price")

ax[1].scatter(Y, Z)
ax[1].set_xlabel("Number of Bathrooms")
ax[1].set_ylabel("House Price")
plt.show()
```
