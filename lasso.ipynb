```python
import numpy as np
import matplotlib.pyplot as plt
from math import ceil
from random import random
n=40
X = np.random.randint(30, 150, (n,1)) # square meters
Y = np.reshape([ceil(x) for [x] in ((X-30) / 120) * 2], (n,1))
def rumor(x):
    if x == 1:
        if random()<0.05: return 2
        return 1
    if random()<0.15: return 1
    return 2
Y = np.reshape([rumor(x) for x in Y], (n,1)) # number of bathrooms
Z = 2500*X + 30000*(Y**2) # price
Z = np.reshape([z+(0.5-random())*70000 for [z] in Z], (n,1))
```

```python
plt.scatter(X,Z)
plt.show()
```

```python
plt.scatter(Y,Z)
plt.show()
```
